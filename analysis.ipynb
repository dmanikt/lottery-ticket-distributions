```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ONLY WORKS FOR LOSS/ACC CURVES
def get_mean_data(run_param, metric):
    trials = [1,2,3,4,5]
    files = ['outputs/mnist/fc1/trial_{}/{}_{}.txt'.format(t,run_param,metric) for t in trials]
    
    data = []
    for f in files:
        data.append(np.loadtxt(f, dtype=float))
    return np.vstack(data).mean(axis=0)
```

```python
run_params = ['0_full_Dinit',
              '6_pruned_Dinit', 
              '6_pruned_Dn',
              #'6_randompruned_Dn',
              #'10_pruned_Dinit',
              #'10_pruned_Dn',
              #'10_randompruned_Dn',
             ]

test_acc_data = {k:get_mean_data(k,'test_acc') for k in run_params}
test_loss_data = {k:get_mean_data(k,'test_loss') for k in run_params}
train_loss_data = {k:get_mean_data(k,'train_loss') for k in run_params}

data = test_acc_data

plt.figure(figsize=(20,10))
for k in run_params:
    plt.plot(data[k], label=k)
    print('{}: best={}, argbest={}'.format(k,data[k].max(),data[k].argmax()))
plt.legend()
plt.show()
```
